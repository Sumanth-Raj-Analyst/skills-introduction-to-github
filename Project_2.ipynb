```python
 import numpy as np
 import pandas as pd
 import matplotlib as plt
 import seaborn as sns
```

```python
df=pd.read_csv("student_data.csv")
df.head()
```

```python
df.drop_duplicates()
```

```python

df['Result'] = df[['MathScore', 'ReadingScore', 'WritingScore']].mean(axis=1)

df

```

```python
out = []

for i in df["PracticeSport"]:
    ch = ""
    if isinstance(i, str):
        for j in i:
            if j in "AEIOUaeiou":
                ch += j.upper()
            else:
                ch += j
    else:
        ch = i
    out.append(ch)

df["PracticeSport"] = out

df
```
