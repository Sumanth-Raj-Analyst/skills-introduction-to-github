#Joins:

1. Inner Join

```python
D1={"ENAME":["A","B","C","D"],"EID":[1,2,3,4],"DEPT_NO":[10,11,11,13]}
df1=pd.DataFrame(D1)
D2={"DEPT_NO":[11,13,13,14],"DEPTNAME":["DEV","DB","DB","ANALYST"]}
df2=pd.DataFrame(D2)

a=pd.merge(df1,df2,on="DEPT_NO",how="inner")
a
a.drop_duplicates()
```

```python
D1={"ENAME":["A","B","C","D"],"EID":[10,12,25,40],"DEPT_NO":[1,2,2,3]}
df1=pd.DataFrame(D1)
D2={"DEPT_NO":[2,3],"DEPTNAME":["DEV","DB"]}
df2=pd.DataFrame(D2)

b=pd.merge(df1,df2,on="DEPT_NO",how="left")
b
```

3. Right Join:

```python
D1={"ENAME":["A","B","C","D"],"EID":[10,12,25,40],"DEPT_NO":[1,2,2,3]}
df1=pd.DataFrame(D1)
D2={"DEPT_NO":[2,3,3,4],"DEPTNAME":["DEV","DB","ANALYST","TESTING"]}
df2=pd.DataFrame(D2)

pd.merge(df1,df2,on="DEPT_NO",how="right")
```

4. Outer Join:

```python
D1={"ENAME":["A","B","C","D"],"EID":[10,12,25,40],"DEPT_NO":[1,2,3,4]}
df1=pd.DataFrame(D1)
D2={"DEPT_NO":[3,4,7,8],"DEPTNAME":["DEV","DB","ANALYST","TESTING"]}
df2=pd.DataFrame(D2)

pd.merge(df1,df2,on="DEPT_NO",how="outer")
```

#Feature Engineering Task:

```python
d1={"Name":["Iron Man","Captain America","Hulk","Black Panter","SpiderMan","Venom","Vision","Thor","Loki","AntMan"],"Maths":[95,90,95,82,85,90,85,86,87,88],"Physics":[90,95,85,80,75,80,85,90,95,80],"Chemistry":[85,89,88,90,95,95,96,97,85,90],"Biology":[90,95,85,87,89,88,85,88,89,90],"Geography":[80,85,89,90,95,97,80,85,90,95]}
df=pd.DataFrame(d1)
df["Per"]=(df["Maths"]+df["Physics"]+df["Chemistry"]+df["Biology"]+df["Geography"])/5
df
Grade=[]

for i in df["Per"]:
  if i>100 or i<0:
    Grade.append("Invalid")
  elif 85<=i<=100:
    Grade.append("A+")
  elif 70<=i<=84.9:
    Grade.append("A")
  elif 50<=i<=69.9:
    Grade.append("B")
  elif 35<=i<=49.9:
    Grade.append("C")
  elif i<=34.9:
    Grade.append("Fail")

df["Grade"]=Grade

df

new_row=pd.DataFrame({"Name":["Mean"], "Maths":[df["Maths"].mean()], "Physics":[df["Physics"].mean()],"Chemistry":[df["Chemistry"].mean()],"Biology":[df["Biology"].mean()],"Geography":[df["Geography"].mean()],"Per":[df["Per"].mean()],"Grade":["WellDone"]})

df=pd.concat([df,new_row],ignore_index=True)

df
```
