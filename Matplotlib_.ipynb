```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

```python
x=np.arange(1,5,1)
y=x**2
z=x**3
print(x)
print(y)

plt.figure(figsize=(5,3))
plt.plot(x,y,color="orange",label="square",marker="*",markersize=15,markerfacecolor="red",markeredgecolor="blue",linestyle="--",linewidth=3)
plt.plot(x,z,color="skyblue",label='cube',marker=".",markersize=15,markerfacecolor="green",markeredgecolor="black",linewidth=5)
plt.grid()
plt.title("Y=MX^2 AND Z=MX^3",fontsize=12)
plt.xlabel("X-AXIX",fontsize=20)
plt.ylabel("Y-AXIX",fontsize=20)
plt.rcParams["font.size"]=14
plt.legend(fontsize=10)


```

```python
arr1=np.arange(1,6,1)
arr2=arr1**2
arr3=arr1**3
plt.subplot(3,2,1)
plt.plot(arr1,arr2)
plt.subplot(3,2,2)
plt.plot(arr1,arr3)
```

```python
x=np.arange(1,11,1)
y=x**2
z=x**3
a=x**4
b=x
c=x**5
print(x)
print(y)
print(z)
print(a)
print(b)
print(c)

plt.figure(figsize=(10,15))
plt.subplot(3,2,1)
plt.title("Population Growth")
plt.plot(x,y,color="orange",label="square",marker=".",markersize=10,markeredgecolor="black",linewidth=3)
for i in range(len(x)):
  plt.annotate(y[i],(x[i],y[i]),ha="right")

plt.gca().set_facecolor("lightblue")
#plt.style.use("dark_backgroud")
plt.xlabel("Population")
plt.ylabel("Year")
plt.legend()
plt.plot(x,z)
plt.subplot(3,2,2)
plt.plot(x,z)
plt.subplot(3,2,3)
plt.plot(x,a)
plt.subplot(3,2,4)
plt.plot(x,b)
plt.subplot(3,2,5)
plt.plot(x,c)
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y,color="red")
plt.plot(x,y,color="green")


plt.figure(figsize=(5,3))
plt.title("Population")
```

```python
var=plt.figure()
var1=var.add_axes([0.1,0.1,0.8,0.8])
x=np.arange(1,6,1)
y=x**2
var1.plot(x,y)
```

```python
arr1=np.arange(1,6,1)
arr2=arr1**2
arr3=arr1**3
fig=plt.figure(figsize=(10,5),facecolor="yellow")
a1=fig.add_axes([0.1,0.1,0.8,0.8])
a1.set_facecolor("lightblue")
a2=fig.add_axes([0.5,0.5,0.4,0.4])

a2.set_facecolor("lightgreen")
a1.plot(arr1,arr2,color="blue",label="50 years")
a1.set_xlabel("Population")
a1.set_ylabel("Years")
a1.legend()
#plt.gca().set_facecolor("lightpink")
a1.set_title("Population Growth for last 50 years",loc="left",fontsize=12)
a2.plot(arr1,arr3,color="orange",label="10 years")
a2.set_xlabel("Population")
a2.set_ylabel("Years")
a2.set_title("Population Growth for last 10 years")
a2.legend()
#plt.gca().set_facecolor("lightblue")
```

```python
fig,axes=plt.subplots(nrows=3,ncols=2)
fig.set_size_inches(10,8)
fig.suptitle("Population")
fig.text(0,0,"Text",ha="right")

axes[1][0].plot(arr1,arr2)
axes[2][1].plot(arr1,arr3)
```
