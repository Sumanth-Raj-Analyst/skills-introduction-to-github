 # import all necessary packages

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

 # Read the sales data csv file


```python
df=pd.read_csv("sales_data.csv",encoding="ISO-8859-1")
df.head()
```

#fetching null data from Amount col


```python
df[df["Amount"].isnull()]
```

#Droping unwanted columns such as staus, unnamed1


```python
df.drop(["Status","unnamed1"],axis=1)
```

#Delete all nan values from the data frame


```python
df.dropna()
df.info()
```

#Rename the column Marital_Status to Marital


```python
d={"Martial_Status":"Marital"}
df.rename(columns={"Marital_Status":"Marital"},inplace=True)
df.head(10)
```

#change the values of Marital col 0-->single, 1--> married


```python
d={0:"Single",1:"Married"}
df["Marital"]=df["Marital"].map(d)
df.head()
```

#change the values of Gender col F-->Female, M-->Male


```python
d1={"F":"Female",'M':"Male"}
df["Gender"]=df["Gender"].map(d1)
df.head()
```

#visualize the count of males and females based on gender column using
seaborn

```python
sns.countplot(x="Gender",data=df,hue="Gender")
```

#group by gender, age_group,
state,Occupation,Product_category,marital_status

```python
a=df.groupby("Gender")["Amount"].sum()

a=pd.DataFrame(a)
b=sns.barplot(x="Gender",y="Amount",hue="Gender",data=a,errorbar=None)
for i in b.containers:
  b.bar_label(i)
```

Conclusion: Females have a higher cumulative spending amount compared to males in the dataset, as evidenced by the bar chart.
The total amount spent by females is approximately 74.34 million, while males spent approximately 31.91 million.
