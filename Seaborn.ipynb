```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("Titanic-Dataset.csv")
df.head()
```

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
#distplot: syntax: sns.distplot(df["col_name"])

plt.figure(figsize=(7,5))
# sns.distplot(df["Age"])

#plotting only histogram:
# sns.distplot(df["Age"],kde=False)

#plotting only kde:
#sns.distplot(df["Age"],hist=False)

#adding bins to the distplot:
# sns.distplot(df["Age"],bins=10,kde=True)

#adding rugplot:
# sns.distplot(df["Age"],bins=10,kde=False,rug=True,hist=False)

#merging all three plots:
# sns.distplot(df["Age"],bins=10,kde=True,rug=True,hist=True)

sns.distplot(df[df["Survived"]==0]["Age"],bins=10,hist_kws={"color":"green"},kde_kws={"color":"darkgreen"},rug=True,rug_kws={"color":"darkblue"})
```

```python
#jointplot:

#plotting scatter plot:
sns.jointplot(x="Survived",y="Age",data=df,kind="scatter")

#plotting kde:
sns.jointplot(x="Survived",y="Age",data=df,kind="kde")
```

```python
#matrix plot:

flights=sns.load_dataset("flights")
flights
```

```python
x=flights.pivot_table(index="year",columns="month",values="passengers")
x
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(x,linewidths=0.5,annot=True,fmt=".1f",cmap="rainbow")
```

```python
sns.clustermap(x,metric="correlation",cmap="coolwarm",annot=True,fmt=".1f",linewidths=0.5,col_cluster=False)
```
