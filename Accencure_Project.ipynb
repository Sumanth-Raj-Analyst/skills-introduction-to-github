```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step1 *Read Data set 1 csv file*

```python
dataset1=pd.read_csv("Content.csv")
dataset1.head()
#dataset1.info()
```

# Step 1 *Removing rows that have values which are missing and duplicate values**

```python
dataset1.dropna()
dataset1.drop_duplicates
dataset1.head()
#dataset1_cleaned.info()
```

# Step 2: Change data types


```python
dataset1["Content ID"]=dataset1["Content ID"].astype(str)
dataset1['Type'] = dataset1['Type'].astype('category')
dataset1['Category'] = dataset1['Category'].astype('category')
print(dataset1.dtypes)
dataset1.head()
```

## *Step 3: Remove irrelevant columns*

```python
dataset1=dataset1.drop(columns=["URL"])
dataset1=dataset1.drop(columns=["User ID"])

dataset1.head()
```

rename column name

```python
dataset1.rename(columns={"Type":"Content Type"},inplace=True)
dataset1.head()
```

# *Cleanig dataset:2*

```python
dataset2=pd.read_csv("ReactionTypes.csv")
dataset2.head()
#dataset2.info()
dataset2.isnull()
```

```python
dataset2=dataset2.rename(columns={"Type":"Reaction Type"})
dataset2.head()
```

# *Cleaning data set 3*

```python
dataset3=pd.read_csv("Reactions.csv")
dataset3.head()
dataset3.isnull()
```

```python
dataset3.drop_duplicates(inplace=True)
dataset3
```

```python
dataset3.dropna(inplace=True)
dataset3
```

```python
dataset3=dataset3.drop(columns=["User ID"])
dataset3
```

```python
dataset3=dataset3.rename(columns={"Type":"Reaction Type"})
dataset3.head()
```

# **DATA MODELING : Merging the tables by concat()**

Step 1: Create the Final Dataset by Merging Tables


```python
merged_data= pd.merge(dataset3, dataset2, on="Reaction Type", how="left")
merged_data
```

```python
data = pd.merge(merged_data, dataset1, on="Content ID", how="left")
data
```

Clean and Verify the Final Dataset


```python
data.dropna()
```

Step 2:*** Calculate the Total Scores for Each Category***

Group by category and calculate the total reaction_score.


```python
category_scores = data.groupby("Category")["Score"].sum().reset_index()
category_scores
```

Sort categories by total scores in descending order

```python
top_5_categories = category_scores.sort_values("Score", ascending=False).head(5)
top_5_categories

```

Reactions in the most popular category

```python
most_popular_category = category_scores.iloc[0]['Category']
total_reactions = data[data['Category'] == most_popular_category].shape[0]
print(f"Reactions to the most popular category ({most_popular_category}): {total_reactions}")
```

```python

plt.figure(figsize=(10, 6))
plt.bar([most_popular_category], [total_reactions], color='orange')

# Titles and Labels
plt.title('Reactions to the Most Popular Category', fontsize=16)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Total Reactions', fontsize=12)
plt.tight_layout()
plt.show()
```

What Was the Month with the Most Posts?

```python
data['Month'] = pd.to_datetime(data['Datetime']).dt.month
posts_per_month = data['Month'].value_counts().reset_index()
posts_per_month.columns = ['Month', 'Post_Count']
most_posts_month = posts_per_month.iloc[0]
print(f"Month with the most posts: {most_posts_month['Month']} with {most_posts_month['Post_Count']} posts")
```

```python
# Visualization
plt.figure(figsize=(10, 6))
posts_per_month.sort_values('Month').plot(
    x='Month',
    y='Post_Count',
    kind='bar',
    color='lightgreen',
    legend=False
)

# Titles and Labels
plt.title('Posts Per Month', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Posts', fontsize=12)
plt.xticks(ticks=range(12), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.tight_layout()
plt.show()
```

# **Top 5 Categories Visualization**

```python
plt.figure(figsize=(10, 6),facecolor="yellow")
sns.barplot(data=top_5_categories, x='Category', y='Score',color='blue')
plt.title('Top 5 Content Categories by Score', fontsize=16)
plt.xlabel('Category', fontsize=5)
plt.ylabel('Total Score', fontsize=5)
plt.xticks(rotation=70)
plt.tight_layout()
plt.show()
```

```python

# Data for the bar chart
categories = ['animals', 'science', 'healthy eating', 'technology', 'food']
scores = [75000, 71000, 69500, 67000, 66000]

# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=scores, y=categories, palette='Blues_d')
plt.title('Top 5 Categories by Aggregate "Popularity" Score', fontsize=14)
plt.xlabel('Aggregate "Popularity" Score', fontsize=12)
plt.ylabel('Categories', fontsize=12)
plt.tight_layout()
plt.show()
```

```python

popularity_percentage = [21.4, 20.3, 19.8, 19.6, 19.0]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
explode = [0.1, 0, 0, 0, 0]  # Highlight the first category
colors = sns.color_palette('pastel')[0:5]
plt.pie(popularity_percentage, labels=categories, autopct='%1.1f%%', startangle=90, explode=explode, colors=colors)
plt.title('Popularity Percentage Share from Top 5 Categories', fontsize=14)
plt.tight_layout()
plt.show()
```
